```r
user_data <- data.frame(
    age = c(25,27,35,41,48,56) ,
    sex = c('M','F','M','F','M','M') ,
    income = c(1500,2500,4000,5500,6300,8000),
    height = c(1.5,1.78,2,1.9,1.4,1.6)
    )
```

```r
print(user_data)
```

```r
sum(user_data$sex == 'M')
```

```r
table(user_data$sex)
```

```r
plot(user_data$age ,user_data$income )
```

```r
lmIncomeAge <- lm(income~age, data = user_data) #Create the linear regression
summary(lmIncomeAge) #Review the results
```

```r
plot(income~age, data = user_data)
abline(lmIncomeAge)
```

```r
cov(user_data$age , user_data$income)
```
